```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
actual_data = pd.read_csv('target_head.csv')
predict_data = pd.read_csv('predict_head.csv')
actual_data = actual_data[:48]
predict_data = predict_data[:48]
```

```python
def drawdata(target, predict):
    index = [i for i in range(len(target))]
    plt.figure(figsize=(12,5))
    plt.plot(target, label = r'Actual', c='#28B463', marker = 'o', linewidth = 1)
    plt.plot(index[len(target) - len(predict):], predict, marker = 'o', label = "Predict", color = 'red', linewidth = 1)
    #plt.ylim(lower, upper)
    plt.legend()
    plt.show()
```

```python
drawdata(actual_data['Barometer'], predict_data['Barometer'])
```
