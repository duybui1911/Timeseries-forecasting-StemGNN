```python
actual_data = pd.read_csv('\output\ECG_data\test\target.csv')
predict_data = pd.read_csv('\output\ECG_data\test\predict.csv')
actual_data = actual_data[:len(actual_data) - len(predict_data) + 48]
predict_data = predict_data[:48]
```

```python
def drawdata(target, predict):
    index = [i for i in range(len(target))]
    plt.figure(figsize=(12,5))
    plt.plot(target, "b-", label = r'Actual', c='#28B463', marker = 'o', linewidth = 1)
    plt.plot(index[len(target) - len(predict):], predict, "r-", marker = 'o', label = "Predict", color = 'red', linewidth = 1)
    #plt.ylim(lower, upper)
    plt.legend()
    plt.show()
```
